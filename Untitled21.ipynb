```python
!pip install keras-tuner
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from keras import Sequential
from keras.layers import Dense, LSTM, Dropout, SimpleRNN, LSTM
from keras.optimizers import Adam, SGD
from keras.callbacks import EarlyStopping
from keras_tuner import RandomSearch, HyperParameters
from sklearn.metrics import mean_squared_error
import math
```

```python
train = pd.read_csv('Google_Stock_Price_Train.csv')
test = pd.read_csv('Google_Stock_Price_Test.csv')
```

```python
train.info()
```

```python
test.info()
```

```python
# Convert to date time type
train['Date'] = pd.to_datetime(train['Date'])
test['Date'] = pd.to_datetime(test['Date'])
```

```python
# Making the Date column as index
train.set_index('Date',drop=True,inplace=True)
test.set_index('Date',drop=True,inplace=True)
```

```python
# Converting the column as string and replacing the comma values with empty string
train['Close'] = train['Close'].str.replace(',', '').astype(float)
```

```python
# Plot the values on matplotlib plot
plt.figure(figsize=(16, 8))
plt.plot(train['Close'], label = "Close")
plt.plot(train['Open'], label = "Open")
plt.plot(train['High'], label = "High")
plt.plot(train['Low'], label = "Low")
plt.title('Closing Price of Google Stock')
plt.xlabel('Year')
plt.ylabel('Closing Price')
plt.legend()
plt.show()
```

Reason for closing price drop by 50% is shown in the url below [link text](https://www.ig.com/au/trading-strategies/google-stock-split-history--what-you-need-to-know-190905.)

```python
# Only using the close column
train = train['Close']
train.head()
```

```python
test = test['Close']
test.head()
```

```python
mm_scalar = MinMaxScaler()
train_data = mm_scalar.fit_transform(np.array(train).reshape(-1,1))
test_data = mm_scalar.transform(np.array(test).reshape(-1,1))
```

PREPARING THE DATA

```python
x_train = []
y_train = []

step = 5
for i in range(step, len(train_data)):
    x_train.append(train_data[i-step:i, 0])
    y_train.append(train_data[i, 0])

x_train, y_train = np.array(x_train), np.array(y_train)

x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
y_train = np.reshape(y_train, (y_train.shape[0], 1))
```

```python
x_train.shape, y_train.shape
```

```python
x_test = []
y_test = []

step = 5
for i in range(step, len(test_data)):
    x_test.append(test_data[i-step:i, 0])
    y_test.append(test_data[i, 0])

x_test, y_test = np.array(x_test), np.array(y_test)

x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
y_test = np.reshape(y_test, (y_test.shape[0], 1))
```

```python
x_test.shape, y_test.shape
```

**EXPERIMENT 1**

  **MODELS**

```python
# RNN
def Vanilla_RNN(hp):
  model = Sequential()
  model.add(SimpleRNN(units = hp.Int('units_1', min_value = 10, max_value = 200, step = 10), activation = 'tanh', input_shape = (10, 1)))
  model.add(Dense(units = hp.Int('units_2', min_value = 10, max_value = 200, step = 10), activation = hp.Choice('act_1', values = ['tanh', 'relu', 'sigmoid'])))
  model.add(Dense(1))
  model.compile(optimizer=Adam(learning_rate=hp.Float('learning_rate', min_value=1e-5, max_value=1e-2, sampling='log')),
                  loss='mse', metrics = ['root_mean_squared_error'])
  return model

tuner_1 = RandomSearch(
    Vanilla_RNN,
    objective='val_loss',
    max_trials=20,
    executions_per_trial=1,
    directory='model_1',
    project_name='model_1'
)

tuner_1.search(x_train, y_train, epochs=50, validation_split=0.2)
```

```python
best_hps = tuner_1.get_best_hyperparameters(num_trials=1)[0]
print(f"Best Units: {best_hps.get('units_1')}")
print(f"Best Units Dense: {best_hps.get('units_2')}")
print(f"Best Activation: {best_hps.get('act_1')}")
print(f"Best Learning Rate: {best_hps.get('learning_rate')}")

# Build the model with the best hyperparameters
best_model = tuner_1.hypermodel.build(best_hps)
```

```python
best_model.summary()
```

```python
history_1 = best_model.fit(
    x_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    validation_batch_size = 32,
    verbose=1
)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(history_1.history['loss'], label='Training Loss')
plt.plot(history_1.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
train_predictions = best_model.predict(x_train)
train_predictions= mm_scalar.inverse_transform(train_predictions)

y_train_plot = np.copy(y_train)
y_train_plot = mm_scalar.inverse_transform(y_train_plot)

plt.figure(figsize=(25, 6))
plt.plot(y_train_plot, label='Actual Values')
plt.plot(train_predictions, label='Predicted Values')
plt.title('Actual vs Predicted Values')
plt.xlabel('Sample Index')
plt.ylabel('Target Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
test_predictions = best_model.predict(x_test)
test_predictions = mm_scalar.inverse_transform(test_predictions)

y_test_plot = np.copy(y_test)
y_test_plot = mm_scalar.inverse_transform(y_test_plot)

plt.figure(figsize=(25, 6))
plt.plot(y_test_plot, label='Actual Values')
plt.plot(test_predictions, label='Predicted Values')
plt.title('Actual vs Predicted Values')
plt.xlabel('Sample Index')
plt.ylabel('Target Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
train_preds = best_model.predict(x_train)
train_preds = mm_scalar.inverse_transform(train_preds)
y_train_rescaled = mm_scalar.inverse_transform(y_train)
train_rmse = math.sqrt(mean_squared_error(y_train_rescaled, train_preds))
print(f"Train RMSE: {train_rmse}")

test_preds = best_model.predict(x_test)
test_preds = mm_scalar.inverse_transform(test_preds)
y_test_rescaled = mm_scalar.inverse_transform(y_test)
test_rmse = math.sqrt(mean_squared_error(y_test_rescaled, test_preds))
print(f"Test RMSE: {test_rmse}")
```
